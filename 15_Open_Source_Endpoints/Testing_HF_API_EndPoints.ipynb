```python
import requests
import os
import getpass

# Configuration parameters
max_new_tokens = 256
top_p = 0.9
temperature = 0.1
model_api_gateway = "https://r13gmxdkn9ixihnp.us-east-1.aws.endpoints.huggingface.cloud"

prompt = "Hello! How are you?"

# Prepare the request payload
json_body = {
    "inputs": prompt,
    "parameters": {
        "max_new_tokens": max_new_tokens,
        "top_p": top_p,
        "temperature": temperature
    }
}
os.environ['HF_TOKEN']= getpass.getpass()
# Set up headers
headers = {
    "Authorization": f"Bearer {os.environ['HF_TOKEN']}",
    "Content-Type": "application/json"
}

# Make the API request
response = requests.post(model_api_gateway, json=json_body, headers=headers)
print(response.json())
```

```python
import nest_asyncio
nest_asyncio.apply()

import asyncio
import aiohttp

async def call_endpoint(model_api_gateway, json_body, headers):
    async with aiohttp.ClientSession() as session:
        async with session.post(model_api_gateway, json=json_body, headers=headers) as response:
            return await response.json()

async def main():
    # Configuration parameters
    max_new_tokens = 256
    top_p = 0.9
    temperature = 0.1
    model_api_gateway = "https://r13gmxdkn9ixihnp.us-east-1.aws.endpoints.huggingface.cloud"
    
    prompt = "Hello! How are you?"
    
    # Prepare the request payload
    json_body = {
        "inputs": prompt,
        "parameters": {
            "max_new_tokens": max_new_tokens,
            "top_p": top_p,
            "temperature": temperature
        }
    }
    
    # Set up headers
    headers = {
        "Authorization": f"Bearer {os.environ['HF_TOKEN']}",
        "Content-Type": "application/json"
    }
    
    # Create multiple concurrent tasks
    tasks = []
    for _ in range(100):
        task = asyncio.create_task(call_endpoint(model_api_gateway, json_body, headers))
        tasks.append(task)
    
    # Execute all tasks concurrently
    results = await asyncio.gather(*tasks)
    for result in results:
        print(result)

# Run the async main function
asyncio.run(main())
```
